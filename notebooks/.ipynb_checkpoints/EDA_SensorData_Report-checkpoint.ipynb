## üìã Summary Statistics

```python
summary = []
for name, df in dataframes.items():
    info = {
        "Station": name,
        "Start": df.index.min(),
        "End": df.index.max(),
        "Total Rows": len(df),
        "Duplicates": df.index.duplicated().sum(),
        "Mean Temperature": df["value_Temperatur"].mean(),
        "Mean PM2.5": df["value_PM2.5"].mean(),
        "Mean PM10": df["value_PM10"].mean(),
    }
    summary.append(info)

pd.DataFrame(summary)
```

## üå°Ô∏è Temperature Over Time

```python
plt.figure(figsize=(14, 6))
for name, df in dataframes.items():
    df['value_Temperatur'].plot(label=name)
plt.title("Temperature Over Time")
plt.xlabel("Timestamp")
plt.ylabel("Temperature (¬∞C)")
plt.legend()
plt.grid(True)
plt.show()
```

## üå´Ô∏è PM2.5 Over Time

```python
plt.figure(figsize=(14, 6))
for name, df in dataframes.items():
    df['value_PM2.5'].plot(label=name)
plt.title("PM2.5 Over Time")
plt.xlabel("Timestamp")
plt.ylabel("PM2.5 (¬µg/m¬≥)")
plt.legend()
plt.grid(True)
plt.show()
```

## üå´Ô∏è PM10 Over Time

```python
plt.figure(figsize=(14, 6))
for name, df in dataframes.items():
    df['value_PM10'].plot(label=name)
plt.title("PM10 Over Time")
plt.xlabel("Timestamp")
plt.ylabel("PM10 (¬µg/m¬≥)")
plt.legend()
plt.grid(True)
plt.show()
```

## üìä Histograms of Sensor Values

```python
for metric in ['value_Temperatur', 'value_PM2.5', 'value_PM10']:
    plt.figure(figsize=(14, 6))
    for name, df in dataframes.items():
        sns.histplot(df[metric], kde=True, label=name, stat="density", bins=50, element="step")
    plt.title(f"Distribution of {metric}")
    plt.xlabel(metric)
    plt.ylabel("Density")
    plt.legend()
    plt.show()
```

## üßØ Missing Value Heatmaps

```python
for name, df in dataframes.items():
    plt.figure(figsize=(14, 2))
    sns.heatmap(df.isnull(), cbar=False)
    plt.title(f"Missing Value Map ‚Äì {name}")
    plt.show()
```

## üîó Correlation Matrices

```python
for name, df in dataframes.items():
    plt.figure(figsize=(10, 8))
    corr = df.corr()
    sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
    plt.title(f"Sensor Correlation Matrix ‚Äì {name}")
    plt.show()
```
